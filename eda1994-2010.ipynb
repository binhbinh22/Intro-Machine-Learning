```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/Users/macbook/Desktop/KHDL project/dataset.csv')
df
```

```python
selected_columns = ['year', 'dengue','Month', 'temperature','rainfall','humidity']
df = df[selected_columns]
df
```

```python
# Thống kê mô tả của dữ liệu
print(df.describe())
```

```python
df['Date'] = pd.to_datetime(df[['year', 'Month']].assign(DAY=1))

plt.figure(figsize=(12, 6))

# Biểu đồ đường cho số ca mắc (dengue)
plt.subplot(4, 1, 1)
sns.lineplot(x='Date', y='dengue', data=df, label='Dengue')
plt.title('Số ca mắc theo từng tháng ')
plt.xlabel('Month')
plt.ylabel('Dengue')
plt.legend()

# Biểu đồ đường cho nhiệt độ (temperature)
plt.subplot(4, 1, 2)
sns.lineplot(x='Date', y='temperature', data=df, label='Temperature')
plt.title('Nhiệt độ theo từng tháng ')
plt.xlabel('Month')
plt.ylabel('Temperature')
plt.legend()

# Biểu đồ đường cho lượng mưa
plt.subplot(4, 1, 3)
sns.lineplot(x='Date', y='rainfall', data=df, label='Rainfall')
plt.title('Lượng mưa theo từng tháng ')
plt.xlabel('Month')
plt.ylabel('rainfall')
plt.legend()

plt.tight_layout()

# Biểu đồ đường cho lượng mưa
plt.subplot(4, 1, 4)
sns.lineplot(x='Date', y='humidity', data=df, label='humidity')
plt.title('Độ ẩm theo từng tháng ')
plt.xlabel('Month')
plt.ylabel('rainfall')
plt.legend()

# Hiển thị biểu đồ
plt.show()
```

```python
# Histoogram
plt.figure(figsize=(10, 6))
plt.hist(df['dengue'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Dengue Cases')
plt.xlabel('Dengue Cases')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.show()
```

```python
# Visualize distribution của 'dengue' và 'temperature'
plt.figure(figsize=(12, 6))

plt.subplot(4, 2, 1)
sns.histplot(df['dengue'], kde=True)
plt.title('Distribution of Dengue ')

plt.subplot(4, 2, 3)
sns.histplot(df['temperature'], kde=True)
plt.title('Distribution of Temperature')

plt.subplot(4, 2, 2)
sns.boxplot(x='dengue', data=df)
plt.title('Boxplot of Dengue ')

plt.subplot(4, 2, 4)
sns.boxplot(x='temperature', data=df)
plt.title('Boxplot of Temperature')

plt.subplot(4, 2, 5)
sns.histplot(df['rainfall'], kde=True)
plt.title('Distribution of Rainfall')

plt.subplot(4, 2, 6)
sns.boxplot(x='rainfall', data=df)
plt.title('Boxplot of Rainfall')

plt.subplot(4, 2, 7)
sns.histplot(df['humidity'], kde=True)
plt.title('Distribution of humidity')

plt.subplot(4, 2, 8)
sns.boxplot(x='humidity', data=df)
plt.title('Boxplot of humidity')

plt.tight_layout()
plt.show()
```

```python
sns.boxplot(data=df['dengue'])

Q1 = df['dengue'].quantile(0.25)
Q3 = df['dengue'].quantile(0.75)
IQR = Q3 - Q1

# Tìm và in ra các điểm ngoại lai
outliers_df = df[(df['dengue'] < (Q1 - 1.5 * IQR)) | (df['dengue'] > (Q3 + 1.5 * IQR))]

print(outliers_df)
```

```python
# Mối quan hệ giữa 'dengue' và 'temperature'
plt.figure(figsize=(10, 6))
sns.scatterplot(x='temperature', y='dengue', data=df)
plt.title('Scatterplot giữa số ca và nhiệt độ')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='rainfall', y='dengue', data=df)
plt.title('Scatterplot giữa số ca và lượng mưa')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='humidity', y='dengue', data=df)
plt.title('Scatterplot giữa số ca và độ ẩm')
plt.show()
```

```python
# Tính ma trận tương quan
correlation_matrix = df[['dengue', 'temperature', 'rainfall','humidity']].corr()

# In ra ma trận tương quan
print("Correlation Matrix:")
print(correlation_matrix)

# Tạo heatmap cho mối tương quan giữa 'dengue', 'temperature', và 'rainfall'
plt.figure(figsize=(10, 8))
heatmap_data = df[['dengue', 'temperature', 'rainfall','humidity']].corr()
sns.heatmap(heatmap_data, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap among Dengue Cases, Temperature, and Rainfall')
plt.show()
```

dengue vs. temperature (0.374539): mối tương quan thuận mạnh mức trung bình giữa số ca dengue và nhiệt độ. Điều này có nghĩa là khi nhiệt độ tăng, số ca dengue cũng có xu hướng tăng, nhưng mối tương quan này không mạnh lắm.

dengue vs. rainfall (0.451685):  mối tương quan thuận mạnh mức trung bình giữa số ca dengue và lượng mưa. Điều này có thể chỉ ra rằng khi lượng mưa tăng, số ca dengue cũng có xu hướng tăng.

dengue vs. humidity (-0.179660): mối tương quan nghịch mạnh mức trung bình giữa số ca dengue và độ ẩm. Điều này có thể ngụ ý rằng có một mối tương quan nghịch giữa độ ẩm và số ca dengue, tức là khi độ ẩm tăng, số ca dengue có xu hướng giảm.

temperature vs. rainfall (0.712357): mối tương quan thuận mạnh giữa nhiệt độ và lượng mưa. Điều này có nghĩa là khi nhiệt độ tăng, cũng có xu hướng lượng mưa tăng.

temperature vs. humidity (0.315997) và rainfall vs. humidity (0.213267): Cả hai mối tương quan này đều là thuận mạnh nhưng mức trung bình. Điều này chỉ ra rằng nhiệt độ và lượng mưa có mối tương quan thuận với độ ẩm, nhưng mối tương quan này không mạnh lắm.
